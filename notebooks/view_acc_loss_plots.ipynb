```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
train_path = 'resources/losses/losses_artist_wikiart_classification_epochs_50_train.pkl'
valid_path = 'resources/losses/losses_artist_wikiart_classification_epochs_50_valid.pkl'
```

```python
with open(train_path, 'rb') as out:
    train_met = pkl.load(out)
with open(valid_path, 'rb') as out:
    valid_met = pkl.load(out)
    
plt.figure(figsize=(16, 9))
plt.plot(list(train_met.values()))
plt.plot(list(valid_met.values()))
plt.show()
```
